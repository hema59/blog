# Explanation

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
```

```python
#creating simulation dataset
X ,Y = make_classification(n_samples = 10, n_features =2, n_redundant =0,n_informative=2, n_clusters_per_class=2, class_sep=1, random_state =2)
```

```python
#visualizing
from matplotlib.colors import ListedColormap
import numpy as np

#datastructure that hold cmap values
cmap_col = ListedColormap(['y','r'])#'#FFFF00', '#00FF00', '#0000FF','#000000'])

plt.figure()
plt.scatter(X[:,0], X[:,1], c=Y, marker='o',s=20, cmap=cmap_col)
plt.title("Figure 1 : Spicy Ramen Plot")
plt.xlabel("Amount of tears")
plt.ylabel("Amount of chilli powder")

plt.legend()
plt.show()
```

```python
#using the sigmoid function
x = np.linspace(-10, 10, 10) 
z = 1/(1 + np.exp(-x)) 

plt.scatter(x=range(0,10),y=Y)
plt.plot(z,c='r')
plt.title("Figure 2 : Probability of Spiciness")
plt.show()
```

```python
#visualizing
from matplotlib.colors import ListedColormap
import numpy as np

#datastructure that hold cmap values
cmap_col = ListedColormap(['y','r'])#'#FFFF00', '#00FF00', '#0000FF','#000000'])

plt.figure()
plt.scatter(X[:,0], X[:,1], c=Y, marker='o',s=20, cmap=cmap_col)
plt.title("Figure 3 : Spicy Ramen : Logistic Regression Model")
plt.xlabel("Amount of tears")
plt.ylabel("Amount of chilli powder")

lreg = LogisticRegression().fit(X,Y)
c = lreg.coef_
i = lreg.intercept_
print(c,i)
#plt.plot(X, coeffecient*X+ intercept,'r' )

plt.plot(X, c[0,1]*X +i,'g')
plt.legend()
plt.show()
```

# Case Study

Let's predict whether a new student's profile is capable enough to get into a university

[data source : https://www.kaggle.com/mohansacharya/graduate-admissions]

```python
import pandas as pd
import os

#finding path, ugh don't ask!
path, file = os.path.split(os.path.abspath("Admission_Predict_Ver1.1.csv"))
path = path+'\data\\'

#loading data
data = pd.read_csv(path+file)
data.shape
```

```python
#drop serial no
data.drop("Serial No.", axis = 1,inplace= True)
def set_binary(p):
    if p>0.6:
        return 1
    else:
        return 0
#yeah, its a little cruel but for the sake of explaining

#convert chances of admit to binary
data['Chance of Admit '] = data['Chance of Admit '].apply(set_binary)
data.head()
```

```python
#this distribution seems okay
data['Chance of Admit '].value_counts()
```

```python
#train-test split
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

features = ['GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR ', 'CGPA',
       'Research']
target = 'Chance of Admit '

X_train, X_test, y_train, y_test = train_test_split(data[features], data[target], random_state = 22, test_size=0.2)
X_train.shape, y_train.shape
```

```python
#build and fit model
log_reg = LogisticRegression().fit(X_train, y_train)

#make predictions
y_pred = log_reg.predict(X_test)

#calculate accuracy
accuracy_score(y_test, y_pred)*100
```

## Linear Regression Explanation

```python
import os

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.datasets.samples_generator import make_blobs
from sklearn.metrics import r2_score, accuracy_score, roc_auc_score

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
#create the regression model
from sklearn.datasets import make_regression

plt.figure()
X, y = make_regression(n_samples = 50, n_features=1,
                            n_informative=1, bias = 100.0,
                            noise = 50, random_state=2)

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state =2)

reg = LinearRegression().fit(X_train, y_train)
coeffecient = reg.coef_

#find regressor intercept
intercept = reg.intercept_
print(coeffecient)
print('Intercept: ', intercept)
plt.xlim(-2.5,3)

plt.plot(X,[20.00]*X+0,'y')
plt.plot(X,[25.00]*X+60,'y')
plt.plot(X,[35.00]*X+150,'y')

plt.plot(X, coeffecient*X+ intercept,'r' )
 
plt.scatter(X, y, marker= 'o', s=50)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Fig 3 : Linear Regression model')

plt.show()
```

## Case Study

This dataset contains parameters collected over the past few years by CDC relating to the life expectancy of people in several parts of the world. For the sake of linear problem, let's consider that this is for one country and proceed ahead.

[Dataset from https://www.kaggle.com/kumarajarshi/life-expectancy-who]

#### As a regression problem, let's try to find the life expectancy of an individual in this set 

```python
#finding path, ugh don't ask!
path, file = os.path.split(os.path.abspath("Life_Expectancy_Data.csv"))
path = path+'\data\\'

#loading data
data = pd.read_csv(path+file)
data.shape
```

```python
data.describe() #might need normalizing later
```

```python
data.head()
```

```python
data.columns
```

```python
target = 'Life expectancy '
cols = ['Life expectancy ',
      'Alcohol', 'percentage expenditure', 'Hepatitis B',
       ' BMI ',  'Polio', 'Total expenditure',
       'Diphtheria ',  'GDP', 
       'Income composition of resources', 'Schooling']
data[cols].corr()
```

```python
data[cols].describe()
```

```python
import missingno as mn
mn.matrix(data[cols])
```

```python
data.dropna(inplace=True)

#splitting data
X_train, X_test, y_train, y_test = train_test_split(data[cols[1:]], data[target], random_state = 2, test_size =0.2)

#create model
reg = LinearRegression().fit(X_train, y_train)

coeffecient = reg.coef_

#find regressor intercept
intercept = reg.intercept_
print('Coeffecient/ weights: ',coeffecient)
print('Intercept/ bias : ', intercept)
```

```python
from sklearn.metrics import mean_squared_error, confusion_matrix
from sklearn.model_selection import validation_curve,cross_val_score

#making predictions
y_pred = reg.predict(X_test)

#validate
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print('Mean Squared Error: {}'.format(mse))
print('R-squared Error: {}'.format(r2))
print('Cross validation Score: ',cross_val_score(reg, X_train, y_train, cv=5))
```

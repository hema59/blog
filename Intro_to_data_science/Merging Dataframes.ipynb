```python
import pandas as pd

Ninjas = {'Name':['Sakura', 'Ino', 'Shikamaru', 'Shisune'],
        'Age':[17, 17, 16, 20]}

data = pd.DataFrame(Ninjas)
data.set_index('Name', inplace= True)
#adding new column
data['Jitsu'] =['Devil Strength','Mind Transfer', 'Shadow Claw','Forensics']

data
```

```python
male ={'Name':['Sasuke', 'Naruto', 'Shikamaru', 'Lee'],
        'Age':[17, 17, 17, 17]}
male = pd.DataFrame(male)
male.set_index('Name', inplace=True)
male
```

```python
#to do outer join
pd.merge(data, male, how = 'outer', left_index = True, right_index =True)
```

```python
#inner join
pd.merge(data, male, how = 'inner', left_index = True, right_index = True)
```

```python
#left join
pd.merge(data, male, how = 'left',left_index = True, right_index = True)
```

```python
#right join
pd.merge(data, male, how = 'right',left_index = True, right_index = True)
```

```python
#merging using column names
data.reset_index(inplace=True)
male.reset_index(inplace= True)
pd.merge(data, male, how='left', left_on='Name', right_on='Name')
```

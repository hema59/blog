```python
import pandas as pd
Ninjas = {'Name':['Sakura', 'Ino', 'Shikamaru', 'Shisune','Naruto','Sasuke','Kakashi'],
        'Age':[17, 17, 16, 20, 17,17,30]}

#create dataframe
data = pd.DataFrame(Ninjas)
data.set_index('Name', inplace= True)

#adding new column
data['Jitsu'] =['Devil Strength','Mind Transfer', 'Shadow Claw','Forensics',
               'Nine-Tails','Sharigan','Copy Ninja']
data
```

```python
#method chaining
data.where(data['Age']==17) \
.reset_index()\
.sort_values(by ='Name')\
.fillna('-')
```

```python
#using apply
def set_status(age):
    if age > 17:
       return 'Kid'
    else:
       return 'Adult'

data['Status'] = data['Age'].apply(set_status)
data
```

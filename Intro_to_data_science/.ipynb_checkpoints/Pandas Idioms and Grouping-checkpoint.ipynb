```python
import pandas as pd
Ninjas = {'Name':['Sakura', 'Ino', 'Shikamaru', 'Shisune','Naruto','Sasuke','Kakashi'],
        'Age':[17, 17, 16, 20, 17,17,30]}

#create dataframe
data = pd.DataFrame(Ninjas)
data.set_index('Name', inplace= True)

#adding new column
data['Jitsu'] =['Devil Strength','Mind Transfer', 'Shadow Claw','Forensics',
               'Nine-Tails','Sharigan','Copy Ninja']
data
```

```python
#method chaining
data.where(data['Age']==17) \
.reset_index()\
.sort_values(by ='Name')\
.fillna('-')
```

```python
#using apply
def set_status(age):
    if age < 17:
       return 'Kid'
    else:
       return 'Adult'

data['Status'] = data['Age'].apply(set_status)
data
```

## Group by

```python
#creating a simple groupby object
groupby_object = data.groupby('Age')
```

```python
#accessing groupby objects in for loop
for group, members in groupby_object:
    print('\nGroup Title: {0}'.format(group), end ='')
    print('\t,Number of rows in group: {0}'.format(len(members)))
    print(members)
```

```python
#using aggregate function along with groupby
import numpy as np
grouped = data.groupby('Status').agg({'Age':np.average})
grouped.Age.round()
```

```python
#level parameter is for multilevel grouping
grouped = data.groupby(level=0)['Age']
for group, members in grouped:
    print('Group Name: {}'.format(group))
    print(members)
```

```python
#chaining wiht groupby
df = data.reset_index()
df.set_index('Jitsu')\
    .groupby(level =0)['Age']\
    .agg({'max':np.max ,'average':np.average})
```
